# Greenways

```python
import pandas as pd
import geopandas
%matplotlib inline
```

```python
# import dataset
steps = pd.read_csv("steps.csv")
# filter to important info
num_steps = steps.groupby("neighborhood").sum()['number_of_steps']
num_steps.sort_values(ascending=False)
```

```python
neighborhoods = geopandas.read_file("Neighborhoods_.shp") # read in the shapefile
neighborhoods.plot() # plot the geopandas dataframe
```

```python
# do the merge
steps_map = neighborhoods.merge(num_steps, how='left', left_on='hood', right_on='neighborhood')
# look at the head to confirm it merged correctly
steps_map[['hood','number_of_steps','geometry']].head()
```

```python
steps_map.plot(column='number_of_steps', # set the data to be used for coloring
               cmap='magma',              # choose a color palette
               edgecolor="white",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Number of Steps"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```

```python
transit_score = steps.groupby("neighborhood").sum()['transit_score']
transit_score.sort_values(ascending=False)
transit_map = neighborhoods.merge(transit_score, how='left', left_on='hood', right_on='neighborhood')
transit_map[['hood','transit_score','geometry']].head()
```

```python
transit_map.plot(column='transit_score', # set the data to be used for coloring
               cmap='viridis',              # choose a color palette
               edgecolor="white",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Transit Score"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
               )
```

```python
top_five_transit = transit_map[['hood', 'transit_score']].sort_values('transit_score', ascending=False).head(5)
top_five_transit
top_five_transit.plot(kind='bar', x="hood", y="transit_score", title="Transit Score by Neighborhood")

# same for number of steps
```
